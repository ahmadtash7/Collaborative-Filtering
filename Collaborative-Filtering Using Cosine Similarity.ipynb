```python
def collaborative_filtering(user_id, user_posts_matrix):
  # perform cosine similarity between the target user and other users
  user_similarities = cosine_similarity(user_posts_matrix)
  
  # get the indices of the top 10 most similar users
  similar_user_indices = user_similarities[user_id].argsort()[::-1][1:]

  # get similarity scores with respect to other users
  similarity_scores = user_similarities[user_id][similar_user_indices]

  recommended_posts = []
  for user_index, sim_score in zip(similar_user_indices, similarity_scores):
      # Posts rated by similar users
      rated_by_similar_user = user_posts_matrix.iloc[user_index]
      # Posts not rated by the target user
      not_rated_by_target_user = (rated_by_similar_user > 0) & (user_posts_matrix.iloc[user_id] == 0)

      # Collect recommendations with similarity scores
      for post_id in user_posts_matrix.columns[not_rated_by_target_user]:
          recommended_posts.append({"post_id": post_id, "similarity_score": sim_score})

      # Stop if we have enough recommendations: we are just keeping top 10 for this example
      if len(recommended_posts) >= 10:
          break

  return recommended_posts
```
